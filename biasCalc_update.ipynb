```python
import pandas as pd
```

```python

url = 'https://raw.githubusercontent.com/jackmheller/modernDataAnalytics/main/Data/ESC-2014-grand_final-full_results.csv'
ESC2014 = pd.read_csv(url, skiprows=1)
ESC2014 = ESC2014.replace(r'\n',0, regex=True)
ESC2014 = ESC2014[['From country', 'To country', 'Points']]
ESC2014['Points'] = ESC2014['Points'].astype(int)
ESC2014
```

```python
df2=ESC2014.groupby(['To country'])['Points'].mean().reset_index()
df2
```

```python
final2014 = ESC2014.merge(df2, on = 'To country', suffixes=('_orig', '_group'))
final2014['Diff'] = final2014['Points_orig'] - final2014['Points_group']
final2014[final2014['From country'] == 'Russia']
```

```python

url = 'https://raw.githubusercontent.com/jackmheller/modernDataAnalytics/main/Data/ESC-2019-grand_final-full_results.csv'
ESC2019 = pd.read_csv(url, skiprows=1)
ESC2019 = ESC2019.replace(r'\n',0, regex=True)
ESC2019 = ESC2019[['From country', 'To country', 'Points']]
ESC2019['Points'] = ESC2019['Points'].astype(int)
df=ESC2019.groupby(['To country'])['Points'].mean().reset_index()
final2019 = ESC2019.merge(df, on = 'To country', suffixes=('_orig', '_group'))
final2019['Diff'] = final2019['Points_orig'] - final2019['Points_group']
final2019[final2019['From country'] == 'Russia']
```

```python
finalBias = final2014.merge(final2015,on=['From country', 'To country'], how = 'outer').merge(final2016, on=['From country', 'To country'], how = 'outer')
finalBias = finalBias.merge(final2018, on=['From country', 'To country'], how = 'outer').merge(final2018, on=['From country', 'To country'], how = 'outer')
finalBias = finalBias.merge(final2019, on=['From country', 'To country'], how = 'outer')
```

```python
diffCols = [4,7,10,13,16,19]
finalBias['True_Bias'] = finalBias.iloc[:,diffCols].mean(axis=1)
```

```python
trueBias = finalBias[['From country', 'To country', 'True_Bias']].sort_values('True_Bias', ascending = False)
trueBias
```

```python
trueBias.to_csv('yearBias.csv')
```

```python
trueBias[trueBias['To country'] == 'Italy']
```

```python
diffCols = [4,7,10,13,16,19]
countDf = finalBias.iloc[:,diffCols]
countDf.columns = range(countDf.columns.size)
countDf[countDf < 0] = -1
countDf[countDf > 0] = 1
countDf[countDf == 0] = None
countDf['Sum'] = countDf.sum(axis=1)
countDf['Count'] = countDf.count(axis=1) - 1
countDf['Bias'] = countDf['Sum'] / countDf['Count']
countDf['From country'] = trueBias['From country']
countDf['To country'] = trueBias['To country']
countDf.sort_values('Bias')
```

```python
countDf.to_csv('countBias.csv')
```
